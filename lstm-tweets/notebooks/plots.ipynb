```python
from pdf2image import convert_from_path
import matplotlib.pyplot as plt
from pathlib import Path

plt.close("all")
dpi = 300

all_fig_paths = list( Path().cwd().glob("**/train_validation_loss.pdf") )
all_fig_paths.sort(key=lambda item: item.parent.stem)
print(f"There are {len(all_fig_paths)} figures.")

fig, axs = plt.subplots(5, 5, figsize=(40, 40))

for ax in axs.flat:
    ax.set_axis_off()

for ax, path in zip(axs.flat, all_fig_paths):
    image = convert_from_path(path, dpi=dpi)[0]
    ax.imshow(image)
    ax.axis("off")

fig.tight_layout()
fig.show()

fig.savefig("all_training_plots.png", dpi=dpi)

```

```python
from pdf2image import convert_from_path
import matplotlib.pyplot as plt
from pathlib import Path
import os

plt.close("all")
dpi = 300

all_fig_paths = list( Path().cwd().glob("**/train_validation_loss.pdf") )
all_fig_paths.sort(key=lambda item: item.parent.stem)
print(f"There are {len(all_fig_paths)} figures.")

for path in all_fig_paths:
    fig, ax = plt.subplots(figsize=(5, 5))
    image = convert_from_path(path, dpi=dpi)[0]
    ax.imshow(image)
    ax.axis("off")

    fig.tight_layout()
    fig.show()

    fig.savefig(os.path.join(path.parent, f"train_validation_loss.png"), dpi=dpi)

```
