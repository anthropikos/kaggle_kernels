```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("../lightning_logs/version_0/metrics.csv")

mask_train_loss = df.train_loss_epoch.notna()
mask_val_loss = df.val_loss.notna()

df1 = df[mask_train_loss]
df2 = df[mask_val_loss]

train_loss = df1.train_loss_epoch
val_loss = df2.val_loss
x = range(len(train_loss))


fig, ax = plt.subplots()

ax.set_title("Train-/Validation-loss over epochs")
ax.set_xlabel("Epoch")
ax.set_ylabel('Avg Binary Cross Entropy (BCE) loss')
ax.scatter(x, train_loss, label='train_loss')
ax.scatter(x, val_loss, label='val_loss')
ax.plot(x, train_loss, alpha=0.3)
ax.plot(x, val_loss, alpha=0.3)
ax.legend()

fig.savefig('figures/Train_val_loss_over_epoch.svg')
```
